```python
class MinMaxNorm():
    def __init__(self, min_val=0, max_val=1):
        self.min_val = min_val
        self.max_val = max_val

    def __call__(self, x):
        if isinstance(x, list):
            return [minmaxNorm(x_i, min_val=self.min_val, max_val=self.max_val) for x_i in x]
        else:
            return minmaxNorm(x, min_val=self.min_val, max_val=self.max_val)


def minmaxNorm(x, min_val=0, max_val=1, global_min=None, global_max=None):
    if global_min and global_max:
        x_min = global_min
        x_max = global_max
    else:
        x_min = x.min()
        x_max = x.max()
    nom = x - x_min
    den = x_max - x_min

    if abs(den) > 1e-4:
            return (max_val - min_val) * (nom / den) + min_val
    else:
        return nom
```

```python
def griffinlim(spectrogram, n_iter = 100, window = 'hann', n_fft = 2048, hop_length = -1, verbose = False):
    if hop_length == -1:
        hop_length = n_fft // 4

    angles = np.exp(2j * np.pi * np.random.rand(*spectrogram.shape))

    t = tqdm(range(n_iter), ncols=100, mininterval=2.0, disable=not verbose)
    for i in t:
        full = np.abs(spectrogram).astype(np.complex) * angles
        inverse = librosa.istft(full, hop_length = hop_length, window = window)
        rebuilt = librosa.stft(inverse, n_fft = n_fft, hop_length = hop_length, window = window)
        angles = np.exp(1j * np.angle(rebuilt))

        if verbose:
            diff = np.abs(spectrogram) - np.abs(rebuilt)
            t.set_postfix(loss=np.linalg.norm(diff, 'fro'))

    full = np.abs(spectrogram).astype(np.complex) * angles
    inverse = librosa.istft(full, hop_length = hop_length, window = window)

    return inverse
```

```python
OMAX, OMIN = 7, 0
# def move_octave(x):
#     ox = int(x[-1])
#     up_rp = np.arange(ox + 1, OMAX + 1)
#     down_rp = np.arange(OMIN, ox)
#     rp = np.hstack([down_rp, up_rp])
#     y = np.random.choice(rp, 1, replace=False)
#     return x[:-1] + str(y[0])

def move_octave(x_pitch, x_ins):
    p_exist = set(pins['pitch'][pins['instrument'] == x_ins].values)
    ox = int(x_pitch[-1])
    if ox + 1 < OMAX:
        upbound = ox+1
    else:
        upbound = OMAX
    if ox - 1 < OMIN:
        lowbound = OMIN
    else:
        lowbound = ox-1
    rp = np.arange(lowbound, upbound+1)
    keep_sample = True
    while keep_sample:
        y = np.random.choice(rp, 1, replace=False)
        if y != ox:
            keep_sample = False
        #if x_pitch[:-1] + str(y[0]) in p_exist:
        #    keep_sample = True
        #else:
        #    keep_sample = False
        keep_sample = False
    return x_pitch[:-1] + str(y[0])
```
